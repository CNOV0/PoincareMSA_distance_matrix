# Calculate the distance in Poincaré

```python
def dist_euclidian(xa, ya, xb, yb):
    return ((xb - xa)**2 + (yb - ya)**2)**0.5
```

```python
dist_poincare = pd.DataFrame(0.0, index=range(252), columns=range(252))

for i in range(len(dist_poincare[0])):
    for j in range(i+1, len(dist_poincare[0])):
        dist_poincare.iloc[i][j] = dist_euclidian(phyl_poincare.iloc[i][0], phyl_poincare.iloc[i][1], phyl_poincare.iloc[j][0], phyl_poincare.iloc[j][1])
        dist_poincare.iloc[j][i] = dist_euclidian(phyl_poincare.iloc[i][0], phyl_poincare.iloc[i][1], phyl_poincare.iloc[j][0], phyl_poincare.iloc[j][1])
print(dist_poincare)
```

# Comparing with distance matrix

```python
# Calculate spearman correlation
correlation, p_value = spearmanr(dist_poincare, phyl)
correlation = pd.DataFrame(correlation)
```
