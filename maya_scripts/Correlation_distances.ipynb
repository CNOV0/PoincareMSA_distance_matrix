# Calculate the distance in PoincarÃ©

```python
def dist_euclidian(xa, ya, xb, yb):
    return ((xb - xa)**2 + (yb - ya)**2)**0.5
```

```python
distance_poincare = pd.DataFrame(0.0, index=range(252), columns=range(252))

for i in range(len(distance_poincare[0])):
    for j in range(i+1, len(distance_poincare[0])):
        distance_poincare.iloc[i][j] = dist_euclidian(dist_poincare.iloc[i][0], dist_poincare.iloc[i][1], dist_poincare.iloc[j][0], dist_poincare.iloc[j][1])
        distance_poincare.iloc[j][i] = dist_euclidian(dist_poincare.iloc[i][0], dist_poincare.iloc[i][1], dist_poincare.iloc[j][0], dist_poincare.iloc[j][1])
```

```python
distance_poincare
```

# Comparing with distance matrix

```python
correlation_matrix = pd.DataFrame(index=dist.columns, columns=distance_poincare.columns)

for col1 in dist.columns:
    for col2 in distance_poincare.columns:
        correlation, _ = spearmanr(dist[col1], distance_poincare[col2])
        correlation_matrix.loc[col1, col2] = correlation
```
