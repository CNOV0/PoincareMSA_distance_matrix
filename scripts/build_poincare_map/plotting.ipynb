```python
import pandas as pd
from pplots import read_embeddings, plot_embedding, rotate

%matplotlib inline
```

Define full path to your folders with embeddings and labels.

```python
path_annotation = '/Users/klanna/UniParis/PoincareMSA/examples/thioredoxins/annotation.csv'
path_embedding = '/Users/klanna/UniParis/results/thioredoxins/fasta0.9/PM3sigma=1.00gamma=1.00cosinepca=0_seed0.csv'
```

Call a function to read this from files.

```python
df = read_embeddings(path_embedding, path_annotation)
```

Specify full name of the output figure.

```python
output_file = '/Users/klanna/UniParis/results/thioredoxins/fasta0.9/species.png'
```

Simple plotting without legend or labels. If you set labels_name=None, it will be a plain embedding without labels. labels_text currently doesn't work.

```python

plot_embedding(
    df,
    labels_name='Species',
    labels_text='all',
    show_text=False,
    title='example',
    file_name=output_file,
    plot_legend=False
    )
```

Rotate the embedding with respect to the root.

```python
df_rot = rotate(df)
```

```python
plot_embedding(
    df_rot,
    labels_name='Species',
    labels_text='all',
    show_text=False,
    title='example',
    file_name=output_file,
    plot_legend=True
    )
```

```python
plot_embedding(
    df_rot,
    labels_name='Species',
    labels_text='all',
    show_text=True,
    title='thioredoxins',
    file_name=output_file,
    plot_legend=False
    )
```

```python
plot_embedding(
    df_rot,
    labels_name='Species',
    labels_text=['root', 'Aeropyrum'],
    show_text=True,
    title='thioredoxins',
    file_name=output_file,
    plot_legend=False
    )
```
